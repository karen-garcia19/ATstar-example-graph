```python
from collections import deque
 
class Graph:
    def __init__(self, adjac_lis):
        self.adjac_lis = adjac_lis
 
    def get_neighbors(self, v):
        return self.adjac_lis[v]
 
   # Esta es una función heurística que tiene valores iguales para todos los nodos
    def h(self, n):
        H = {
            'A': 1,
            'B': 1,
            'C': 1,
            'D': 1
        }
 
        return H[n]
 
    def a_star_algorithm(self, start, stop):
        # En este open_lst hay una lista de nodos que se han visitado, pero quién
        # los vecinos no siempre han sido inspeccionados, comienza con el comienzo 
  #nodo
        # Y closed_lst es una lista de nodos que se han visitado
         # y quiénes son los vecinos que siempre han sido inspeccionados
        open_lst = set([start])
        closed_lst = set([])
 
        # poo tiene distancias presentes desde el inicio hasta todos los demás nodos
         # el valor predeterminado es + infinito
        poo = {}
        poo[start] = 0
 
        # par contiene un mapeo adjac de todos los nodos
        par = {}
        par[start] = start
 
        while len(open_lst) > 0:
            n = None
 
            # encontrará un nodo con el valor más bajo de f () -
            for v in open_lst:
                if n == None or poo[v] + self.h(v) < poo[n] + self.h(n):
                    n = v;
 
            if n == None:
                print('Path does not exist!')
                return None
 
            # si el nodo actual es la parada
             # luego comenzamos de nuevo desde el principio
            if n == stop:
                reconst_path = []
 
                while par[n] != n:
                    reconst_path.append(n)
                    n = par[n]
 
                reconst_path.append(start)
 
                reconst_path.reverse()
 
                print('Path found: {}'.format(reconst_path))
                return reconst_path
 
           # para todos los vecinos del nodo actual hacer
            for (m, weight) in self.get_neighbors(n):
              # si el nodo actual no está presente tanto en lst abierto como en lst cerrado
                # agréguelo al lst abierto y tenga en cuenta n como par
                if m not in open_lst and m not in closed_lst:
                    open_lst.add(m)
                    par[m] = n
                    poo[m] = poo[n] + weight
 
                # de lo contrario, compruebe si es más rápido visitar primero n, luego m
                 # y si es así, actualice los datos par y poo
                 # y si el nodo estaba en closed_lst, muévelo a open_lst
                else:
                    if poo[m] > poo[n] + weight:
                        poo[m] = poo[n] + weight
                        par[m] = n
 
                        if m in closed_lst:
                            closed_lst.remove(m)
                            open_lst.add(m)
            # eliminar n de open_lst y agregarlo a closed_lst
             # porque todos sus vecinos fueron inspeccionados
            open_lst.remove(n)
            closed_lst.add(n)
 
        print('Path does not exist!')
        return None
```

```python
adjac_lis = {
    'A': [('B', 1), ('C', 3), ('D', 7)],
    'B': [('D', 5)],
    'C': [('D', 12)]
}
graph1 = Graph(adjac_lis)
graph1.a_star_algorithm('A', 'D')
```
